```python
def d_1(s,T,t,K,x,r):
    return(1/(s*math.sqrt(T-t))*(math.log(x/K)+(r+s**2/2)*(T-t)))
def d_2(s,T,t,K,x,r):
    return(d_1(s,T,t,K,x,r)-s*math.sqrt(T-t))
```

```python
def Phi(d):
    res, err = quad(lambda x: math.exp(-x**2/2),-math.inf,d)
    return(1/(math.sqrt(2*math.pi))*res)
```

```python
def V_BS(s,T,t,K,x,r):
    return(x*Phi(d_1(s,T,t,K,x,r))-K*math.exp(-r*(T-t))*Phi(d_2(s,T,t,K,x,r)))

```

```python
def g_(s,T,t,K,x,r,V_mkt):
    return(V_BS(s,T,t,K,x,r)-V_mkt)
```

```python
def gPrime(s,T,t,K,x,r, vMrkt, h):
    return((g_(s+h,T,t,K,x,r,vMrkt)-g_(s,T,t,K,x,r,vMrkt))/h)



print(g_(1.6,3,0,1.5,1.2,1,V_BS(0.6,3,0,1.5,1.2,1)))
```

```python
def methode_Brent(s0,s1,s2,epsilon,T,t,K,x,r,V_mkt):
    sigma=[]
    sigma.append(s0)
    sigma.append(s1)
    sigma.append(s2)
    k=2
    while abs(sigma[k]-sigma[k-1])>=epsilon :
            q1=sigma[k] * g_(sigma[k-1],T,t,K,x,r,V_mkt)*g_(sigma[k-2],T,t,K,x,r,V_mkt) 
            q2=((g_(sigma[k],T,t,K,x,r,V_mkt)-g_(sigma[k-1],T,t,K,x,r,V_mkt))*(g_(sigma[k],T,t,K,x,r,V_mkt)-g_(sigma[k-2],T,t,K,x,r,V_mkt)) )
            q3=(sigma[k-1])* g_(sigma[k-2],T,t,K,x,r,V_mkt)*g_(sigma[k],T,t,K,x,r,V_mkt) 
            q4=(g_(sigma[k-1],T,t,K,x,r,V_mkt)-g_(sigma[k-2],T,t,K,x,r,V_mkt))*(g_(sigma[k-1],T,t,K,x,r,V_mkt)-g_(sigma[k],T,t,K,x,r,V_mkt)) 
            q5=(sigma[k-2]) * g_(sigma[k-1],T,t,K,x,r,V_mkt)*g_(sigma[k],T,t,K,x,r,V_mkt) 
            q6=(g_(sigma[k-2],T,t,K,x,r,V_mkt)-g_(sigma[k-1],T,t,K,x,r,V_mkt))*(g_(sigma[k-2],T,t,K,x,r,V_mkt)-g_(sigma[k],T,t,K,x,r,V_mkt)) 
            if(q2==0 or q4==0 or q6==0):
                break
            if (abs((q1/q2  +  q3/q4  +  q5/q6 )-sigma[k])<=10e-6) :
                sigma.append(sigma[k-1]-g_(sigma[k-1],T,t,K,x,r,V_mkt)*(sigma[k-1]-sigma[k-2])/(g_(sigma[k-1],T,t,K,x,r,V_mkt)-g_(sigma[k-2],T,t,K,x,r,V_mkt)))
            else :
                sigma.append(q1/q2  +  q3/q4  +  q5/q6 )
            k=k+1
    return( sigma[k] )



methode_Brent(1,1.3,1.5,10e-6,1,0,1,1,1,V_BS(0.16,1,0,1,1,1))
```

```python
def newton_raphson (s0,N,T,t,K,x,r,V_mkt,h):
    sigma=[]
    sigma.append(s0)
    sigma.append(sigma[0]-(g_(sigma[0],T,t,K,x,r,V_mkt))/gPrime(sigma[0],T,t,K,x,r,V_mkt,h))
    k=1
    while (abs((sigma[k]-sigma[k-1]))>10**-6 and k<=N):
        k+=1
        sigma.append(sigma[k-1]-(g_(sigma[k-1],T,t,K,x,r,V_mkt)/gPrime(sigma[k-1],T,t,K,x,r,V_mkt,h)))
    return(sigma)


newton_raphson(1,1000,1,0,1,1,1,V_BS(0.16,1,0,1,1,1),0.0001)
```

```python
def bissection(g,a,b,epsilon,T,t,K,x,r, vMrkt):
    left=a
    right=b
    m = (a+b)/2
    while(g(right,T,t,K,x,r, vMrkt)-g(left,T,t,K,x,r, vMrkt)>epsilon):
        if(g(m,T,t,K,x,r, vMrkt)<0):
            left=m
        else:
            right=m
        m = (left+right)/2
    return(m)

vmktTest = V_BS(0.6,1,0,1,1,1)
l=[0.1]
for i in range(2,100):
    l.append(i*0.1)
    
gLs = []
for i in range(0,99):
    gLs.append(g_(l[i],1,0,1,1,1,vmktTest))
    
plt.plot(l, gLs)
bissection(g_,0.001,10,0.001,1,0,1,1,1,vmktTest)
```

```python
def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true)))
modelGEN.compile(loss = root_mean_squared_error, optimizer = "adam")

modelGEN.fit(x_train_transform, y_train_transform, batch_size=32,validation_data = (x_test_transform,y_test_transform),epochs = 100, verbose = True,shuffle=1)#,callbacks=[earlystop])
```
